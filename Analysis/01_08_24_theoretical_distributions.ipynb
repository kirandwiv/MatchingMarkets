```python
import numpy as np
import sympy as sp
import math
```

```python
def get_p(k: float):
    p = sp.Symbol('p')
    return sp.nsolve((1-p)**k - np.e**(-1*(1-(1-p)**k)/p), 0.5)

def mu(p,k): return (1/p)*(1 - (1-p)**k)
def unmatched(p,k): return (1-p)**k
def underdemanded(p,k): return (1 + mu(p,k)) * math.exp( - mu(p,k))

```
